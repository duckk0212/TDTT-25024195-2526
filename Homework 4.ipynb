# W4A16 
In các số chẵn chia hết cho 3 bé hơn 100

```python
print('Các số chẵn chia hết cho 3 bé hơn 100 là:')
for i in range(1, 100):
    if i % 2 == 0 and i % 3 == 0:
        print(i, end=' ')
```

# W4A18 
Nhập vào số nguyên dương a và b, in toàn bộ ước chung của a và b

```python
a = int(input('Nhập số nguyên dương a: '))
b = int(input('Nhập số nguyên dương b: '))
limit = a if a < b else b
print(f'Các ước chung của {a} và {b} là:')
for i in range(1, limit + 1):
    if a % i == 0 and b % i == 0:
        print(i, end=' ')
```

# W4A20
Nhập vào số nguyên dương n, kiểm tra xem n có phải là số dạng 2^k hay không

```python
n = int(input('Nhập n: '))
temp = n
if n <= 0:
    check = False
else:
    while n % 2 == 0:
        n //= 2
    check = (n == 1)
print(f'{temp} là số dạng 2^k' if check else f'{temp} không phải dạng 2^k')
```

# W4A21 
Nhập vào số nguyên dương n, tính tổng các chữ số của n

```python
n = int(input('Nhập n: '))
tong = 0
temp = n
while n > 0:
    tong += n % 10
    n //= 10
print(f'Tổng các chữ số của {temp} là: {tong}')
```

# W4A22 
Nhập vào số nguyên dương n, đếm chữ số chẵn và lẻ

```python
n = int(input('Nhập n: '))
chan = 0
le = 0
if n == 0: chan = 1
while n > 0:
    if (n % 10) % 2 == 0: chan += 1
    else: le += 1
    n //= 10
print(f'Chẵn: {chan}, Lẻ: {le}')
```

# W4A23 
Nhập n, tìm k sao cho S(k) = 1+2+...+k lớn nhất nhưng nhỏ hơn n

```python
n = int(input('Nhập n: '))
s_k = 0
k = 0
while s_k + (k + 1) < n:
    k += 1
    s_k += k
print(f'k lớn nhất là: {k}, S({k}) = {s_k}')
```

# W4A24 
Nhập vào A, tìm n nhỏ nhất sao cho 1 + 1/2 + ... + 1/n > A

```python
A = float(input('Nhập A: '))
tong = 0
n = 0
while tong <= A:
    n += 1
    tong += 1/n
print(f'n nhỏ nhất là: {n}')
```

# W4A25 
Nhập dãy số nguyên, ngưng khi nhập -1. In số lớn nhất, nhỏ nhất

```python
max_val = min_val = None
while True:
    num = int(input('Nhập số (-1 để dừng): '))
    if num == -1: break
    if max_val is None or num > max_val: max_val = num
    if min_val is None or num < min_val: min_val = num
print(f'Max: {max_val}, Min: {min_val}')
```

# W4A26 [Fibonaci1] 
Tìm số Fibonacci lớn nhất không vượt quá A

```python
A = int(input('Nhập A: '))
f1, f2 = 1, 1
result = 1
while f2 <= A:
    result = f2
    f1, f2 = f2, f1 + f2
print(f'Số Fibonacci lớn nhất không quá {A} là: {result}')
```

# W4A27 
Đếm số từ (Dùng for trên index; KHÔNG dùng hàm split(), len())

```python
s = input('Nhập chuỗi: ')
length = 0
for char in s: length += 1
count = 0
in_word = False
for i in range(length):
    if s[i] != ' ':
        if not in_word:
            count += 1
            in_word = True
    else: in_word = False
print(f'Số từ: {count}')
```

# W4A28 
In từ đầu tiên (Dùng for trên index; KHÔNG dùng hàm split())

```python
s = input('Nhập chuỗi: ')
length = 0
for char in s: length += 1
first_word = ''
started = False
for i in range(length):
    if s[i] != ' ':
        first_word += s[i]
        started = True
    elif started: break
print(f'Từ đầu tiên: {first_word}')
```

# W4A29 
Đổi ký tự hoa thành thường và thường thành hoa

```python
s = input('Nhập chuỗi: ')
res = ''
for char in s:
    if 'a' <= char <= 'z': res += char.upper()
    elif 'A' <= char <= 'Z': res += char.lower()
    else: res += char
print(f'Kết quả: {res}')
```

# W4A30 
Nhập chuỗi 3 số nguyên cách nhau bởi dấu phẩy, tính tổng

```python
s = input('Nhập 3 số (vd: 3, 12, 15): ')
parts = s.split(',')
tong = 0
for p in parts: tong += int(p.strip())
print(f'Tổng: {tong}')
```

# W4A31 
Đếm số ký tự in hoa, in thường, ký tự số

```python
s = input('Nhập chuỗi: ')
hoa = thuong = so = 0
for c in s:
    if 'A' <= c <= 'Z': hoa += 1
    elif 'a' <= c <= 'z': thuong += 1
    elif '0' <= c <= '9': so += 1
print(f'Hoa: {hoa}, Thường: {thuong}, Số: {so}')
```
