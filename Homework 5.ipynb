Viết hàm kiểm tra một số có phải số Amstronhg hay không? Hàm trả về kiểu bool, có tham số là một số nguyên x

```python
def is_Armstrong_number(n):
    original = n
    x = len(str(n))
    sum = 0
    while n > 0:
        a = n % 10
        sum += a**x
        n = n // 10
    return sum == original

print(is_Armstrong_number(153))
print(is_Armstrong_number(123)) 
```

# W5A18
VIết hàm kiểm tra một số có phải số đối xứng không, xử lý trên dữ liệu kiểu số nguyên.

```python
def is_Palindrome_number(n):
    digit = []
    while n>0:
        digits = n%10
        digit.append(digits)
        n = n//10

    digit.reverse()

    left = 0
    right = len(digit) - 1

    while left < right:
        if digit[left] != digit[right]:
            return False
        left += 1
        right -= 1

    return True

print(is_Palindrome_number(12321))
print(is_Palindrome_number(12345))
```

# W5A22
Viết hàm sum_halves(s) nhận chuỗi 6 chữ số và trả về tổng của hai số 3 chữ số đầu và 3 chữ số cuối.

```python
def sum_halves(s):
    if len(str(s)) < 6:
        return "String must be at least 6 characters long."
    else:
        digits = []
        while s > 0:
            a = s % 10
            digits.append(a)
            s = s // 10
        digits.reverse()
        sum1 = digits[0]*100 + digits[1]*10 + digits[2]
        sum2 = digits[3]*100 + digits[4]*10 + digits[5]
        return sum1 + sum2
print(sum_halves(123456))
print(sum_halves(12345))
```

```python
def are_collinear(x1, y1, x2, y2, x3, y3):
    AB = ((x2 - x1)**2 + (y2 - y1)**2)**0.5
    AC = ((x3 - x1)**2 + (y3 - y1)**2)**0.5
    BC = ((x3 - x2)**2 + (y3 - y2)**2)**0.5
    p = (AB + AC + BC) / 2
    if p*(p-AB)*(p-AC)*(p-BC) == 0:
        return True
    else:
        return False
print(are_collinear(1, 2, 2, 3, 3, 4))
print(are_collinear(1, 2, 2, 3, 3, 5))
```

# W5A25
Viết hàm point_in_quadrilateral(quad, p) kiểm tra điểm p có nằm trong hoặc trên biên của tứ giác quad hay không. Trong đó: quad là một tứ giác (4 đỉnh – 4 điểm), p là một điểm cần kiểm tra. VD: quad = [(0,0), (4,0), (4,3), (0,3)], p = (2,1)

```python
def point_in_quadrilateral(quad, p):
    xp, yp = p
    (x1, y1), (x2, y2), (x3, y3), (x4, y4) = quad

    # --- FIX 1 + 2 + 3: kiểm tra 3 điểm thẳng hàng (dùng nhân chéo, không chia)
    def collinear(xa, ya, xb, yb, xc, yc):
        return (yb - ya)*(xc - xa) == (yc - ya)*(xb - xa)

    if collinear(x1,y1,x2,y2,x3,y3) or \
       collinear(x1,y1,x2,y2,x4,y4) or \
       collinear(x1,y1,x3,y3,x4,y4) or \
       collinear(x2,y2,x3,y3,x4,y4):
        return "The quadrilateral is not valid."

    # --- giữ nguyên cách tính cạnh
    AB = ((x2 - x1)**2 + (y2 - y1)**2)**0.5
    AC = ((x3 - x1)**2 + (y3 - y1)**2)**0.5
    AD = ((x4 - x1)**2 + (y4 - y1)**2)**0.5
    BC = ((x3 - x2)**2 + (y3 - y2)**2)**0.5
    CD = ((x4 - x3)**2 + (y4 - y3)**2)**0.5

    # --- diện tích tứ giác (chia 2 tam giác)
    P_ABC = (AB + AC + BC) / 2
    S_ABC = (P_ABC*(P_ABC-AB)*(P_ABC-AC)*(P_ABC-BC))**0.5

    P_ACD = (AC + AD + CD) / 2
    S_ACD = (P_ACD*(P_ACD-AC)*(P_ACD-AD)*(P_ACD-CD))**0.5

    S_ABCD = S_ABC + S_ACD

    # --- khoảng cách từ P
    PA = ((xp - x1)**2 + (yp - y1)**2)**0.5
    PB = ((xp - x2)**2 + (yp - y2)**2)**0.5
    PC = ((xp - x3)**2 + (yp - y3)**2)**0.5
    PD = ((xp - x4)**2 + (yp - y4)**2)**0.5

    # --- diện tích 4 tam giác nhỏ
    P_PAB = (PA + AB + PB) / 2
    S_PAB = (P_PAB*(P_PAB-PA)*(P_PAB-AB)*(P_PAB-PB))**0.5

    P_PBC = (PB + BC + PC) / 2
    S_PBC = (P_PBC*(P_PBC-PB)*(P_PBC-BC)*(P_PBC-PC))**0.5

    P_PCD = (PC + CD + PD) / 2
    S_PCD = (P_PCD*(P_PCD-PC)*(P_PCD-CD)*(P_PCD-PD))**0.5

    P_PAD = (PA + AD + PD) / 2
    S_PAD = (P_PAD*(P_PAD-PA)*(P_PAD-AD)*(P_PAD-PD))**0.5

    S = S_PAB + S_PBC + S_PCD + S_PAD

    # --- FIX 4: so sánh float bằng sai số
    eps = 0.000001

    if S_ABCD - eps <= S <= S_ABCD + eps:
        return True
    else:
        return False
    
print(point_in_quadrilateral([(0,0),(4,0),(4,4),(0,4)],(2,2)))
print(point_in_quadrilateral([(0,0),(4,0),(4,4),(0,4)],(5,5)))
```

```python
def point_in_quadrilateral(quad, p):
    def cross(a, b, c):
        return (b[0]-a[0])*(c[1]-a[1]) - (b[1]-a[1])*(c[0]-a[0])

    signs = []
    for i in range(4):
        a = quad[i]
        b = quad[(i+1)%4]
        signs.append(cross(a, b, p))

    return all(s >= 0 for s in signs) or all(s <= 0 for s in signs)

print(point_in_quadrilateral([(0,0),(4,0),(4,4),(0,4)],(2,2)))
print(point_in_quadrilateral([(0,0),(4,0),(4,4),(0,4)],(5,5)))
```

# W5A28
Viết hàm tính góc ∠ABC (độ) từ tọa độ A(x1,y1), B(x2,y2), C(x3,y3)

```python
def tinh_goc(x1, y1, x2, y2, x3, y3):
    import math
    AB = ((x2 - x1)**2 + (y2 - y1)**2)**0.5
    AC = ((x3 - x1)**2 + (y3 - y1)**2)**0.5
    BC = ((x3 - x2)**2 + (y3 - y2)**2)**0.5
    cos_B = (AB**2 + BC**2 - AC**2) / (2*AB*BC)
    angle_B = math.acos(cos_B) * 180 / math.pi
    return angle_B
print(tinh_goc(0, 0, 1, 0, 1, 1))
print(tinh_goc(0, 0, 1, 0, 0, 1))
```

# W5A33
Viết hàm smart_taxi_fare(car_type, start_hour, km, wait_minutes) tính tổng tiền taxi theo quy tắc giá mở cửa, giá km, phí chờ, phụ phí ban đêm và giảm giá.


```python
def smart_taxi_fare(car_type, start_hour, km, wait_minutes):
    
    # giá cơ bản
    if car_type == "standard":
        base = 10000
        per_km = 12000
    elif car_type == "premium":
        base = 15000
        per_km = 15000
    else:
        return "Invalid car type"

    # tiền quãng đường
    fare = base + km * per_km

    # phí chờ
    fare += wait_minutes * 500

    # phụ phí ban đêm
    if start_hour >= 22 or start_hour < 5:
        fare *= 1.2

    # giảm giá
    if km > 20:
        fare *= 0.9

    return int(fare)

print(smart_taxi_fare("standard", 23, 25, 10))
print(smart_taxi_fare("premium", 14, 15, 5))
```

# W5A34
Viết hàm sum_1_to_n(n) trả về tổng các số x từ 1 đến n, nếu x có các chữ số đôi một khác nhau. Nếu n ≤ 0 trả về 0.

```python
def sum_1_to_n(n):
    for i in range(1, n+1):
        if len(str(n)) == len(set(str(n))):
            return sum(range(1, n+1))
        else:
            return "n must have all distinct digits"
print(sum_1_to_n(10))
print(sum_1_to_n(11))
```
