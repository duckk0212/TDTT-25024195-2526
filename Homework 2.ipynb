## Bài 1: Nhập vào 2 cạnh của một hình chữ nhật. In ra màn hình diện tích và chu vi của nó. 

```python
chieudai = int(input())
chỉeurong = int(input())
chuvi = (chieudai + chỉeurong) * 2
dientich = chieudai * chỉeurong
print("Chu vi hình chữ nhật là:", chuvi)
print("Diện tích hình chữ nhật là:", dientich)
```

### Bài 3 : Tính tam giác: Nhâp vào ba số a,b,c bất kì. Kiểm tra xem nó có thể là độ dài ba cạnh hay không, nếu không thì in ra màn hình 'Khong la ba canh cua mot tam giac'. Ngược lại, kiểm tra xem nó là tam giác gì: Đều, Cân, Vuông, hay Tam giác thường. Tính diện tích, chu vi của tam giác và in ra mànhình.

```python
a,b,c = map(int, input().split())
if not a + b > c or not a + c > b or not b + c > a:
    print("Khong la ba canh cua mot tam giac")
else:
    if a == b == c:
        print("Tam giac deu")
    elif a == b or a == c or b == c:
        print("Tam giac can")
    elif a**2 + b**2 == c**2 or a**2 + c**2 == b**2 or b**2 + c**2 == a**2:
        print("Tam giac vuong")
    else:
        print("Tam giac thuong")
    chuvi = a + b + c
    print("Chu vi tam giac la:", chuvi)
    p = chuvi / 2
    dientich = (p*(p-a)*(p-b)*(p-c)) ** 0.5
    print("Dien tich tam giac la:", dientich)
    
```

## Bài 5: Viết chương trình giải phương trình bậc 2 (ax2 + bx + c = 0)

```python
a,b,c = map(int, input().split())
delta = b**2 - 4*a*c
if delta < 0:
    print("Phuong trinh vo nghiem")
elif delta == 0:
    nghiem = -b/(2*a)
    print("Phuong trinh co nghiem kep:", nghiem)
else:
    nghiem1 = (-b + delta**0.5) / (2*a)
    nghiem2 = (-b - delta**0.5) / (2*a)
    print("Phuong trinh co hai nghiem phan biet:")
    print("Nghiem 1:", nghiem1)
    print("Nghiem 2:", nghiem2)
```

## Bài 8: Giải hệ phương trình tuyến tính:
### ax + by = m

### cx + dy = n

```python
a,b,c,d,m,n = map(float, input().split())
y = (a*n - c*m) / (a*d - b*c)
x = (d*m - b*n) / (a*d - b*c)
if a*d - b*c == 0:
    if a*n - c*m == 0 and d*m - b*n == 0:
        print("He phuong trinh co vo so nghiem")
    else:
        print("He phuong trinh vo nghiem")
else:
    print("Nghiem cua he phuong trinh la:")
    print("x =", x)
    print("y =", y)
```

## Bài 9: Nhập vào thời gian 1 công việc nào đó là x giây. Hãy chuyển đổi và viết ra màn hình số thời gian trên dưới dạng bằng bao nhiêu giờ, bao nhiêu phút, bao nhiêu giây. 

```python
x = int(input())
giờ = x // 3600
phut = (x % 3600) // 60
giay = x % 60
print("Thoi gian la: {} gio {} phut {} giay".format(giờ, phut, giay))
```
