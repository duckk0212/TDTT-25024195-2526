## Bài 1: In các số dương lẻ bé hơn 100 


```python
for i in range (1, 101):
    if i % 2 != 0:
        print(i)
```

## Bài 2: In các số dương chẵn chia hết cho 3 bé hơn 100 

```python
for i in range (1, 101):
    if i % 2 == 0 and i % 3 == 0:
        print(i)
```

## Bài 3: Nhập vào số nguyên dương a, in ra bảng cửu chương của a 

```python
a = int(input())
for i in range (1, 11):
    if i < 11:
        print(f"{a} x {i} = {a*i}")
```

# Bài 4
Nhập h, in tam giác cân

```python
h = int(input())

for i in range(1, h+1):
    for j in range(h-i):
        print(" ", end="")
    if i == 1:
        print("*")
    elif i == h:
        print("*"*(2*h-1))
    else:
        print("*" + " "*(2*i-3) + "*")
```

# Bài 8
In ước chung của a và b

```python
a = int(input())
b = int(input())

for i in range(1, min(a,b)+1):
    if a % i == 0 and b % i == 0:
        print(i)
```

# Bài 9
Kiểm tra số chính phương

```python
a = int(input())
i = 1

while i*i <= a:
    if i*i == a:
        print("YES")
        break
    i += 14
else:
    print("NO")
```

# Bài 10
Kiểm tra số hoàn hảo

```python
a = int(input())
s = 0

for i in range(1, a):
    if a % i == 0:
        s += i

print("YES" if s == a else "NO")
```

# Bài 14

```python
x = int(input())
maxx = x
minn = x

while True:
    x = int(input())
    if x == -1:
        break
    if x > maxx:
        maxx = x
    if x < minn:
        minn = x

print(maxx, minn)
```

# Bài 16

```python
n = int(input())
chan = le = 0

while n > 0:
    d = n % 10
    if d % 2 == 0:
        chan += 1
    else:
        le += 1
    n //= 10

print(chan, le)
```
